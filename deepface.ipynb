```python
from deepface import DeepFace
```

```python
ime_file = "/content/drive/MyDrive/ai/qwrqw.jpg"
```

```python
import cv2
```

```python
img = cv2.imread(ime_file)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(img)
```

```python
try:
  emotions = DeepFace.analyze(img,actions=["emotion"])
  ages = DeepFace.analyze(img,actions=["age"])
  races = DeepFace.analyze(img,actions=["race"])
  genders = DeepFace.analyze(img,actions=["gender"])
except:
   pass
```

```python
age = ages[0]['age']
age
```

```python
emotion = emotions[0]['dominant_emotion']
emotion
```

```python
race =races[0]['dominant_race']
race
```

```python
gender =genders[0]['dominant_gender']
gender
```

```python
e_d = {'angry':"" ,
   'disgust':"",
   'fear':"",
   'happy':"",
   'sad':"",
   'surprise':"",
   'neutral':""}
```

```python
r_d ={'asian':"",
   'indian':"",
   'black':"",
   'white':"",
   'middle eastern':"",
   'latino hispanic':""}
```

```python
g_d = {'Woman':"", 'Man':""}
```

```python
text = f": {age} \n" + f": {g_d[gender]} \n" + f": {e_d[emotion]} \n" + f": {r_d[race]}" 
print(text)
```
