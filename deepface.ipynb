```python
from deepface import DeepFace
```

```python
ime_file = "/content/drive/MyDrive/ai/qwrqw.jpg"
```

```python
import cv2
```

```python
img = cv2.imread(ime_file)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(img)
```

```python
try:
  emotions = DeepFace.analyze(img,actions=["emotion"])
  ages = DeepFace.analyze(img,actions=["age"])
  races = DeepFace.analyze(img,actions=["race"])
  genders = DeepFace.analyze(img,actions=["gender"])
except:
   pass
```

```python
age = ages[0]['age']
age
```

```python
emotion = emotions[0]['dominant_emotion']
emotion
```

```python
race =races[0]['dominant_race']
race
```

```python
gender =genders[0]['dominant_gender']
gender
```

```python
e_d = {'angry':"生氣" ,
   'disgust':"噁心",
   'fear':"害怕",
   'happy':"開心",
   'sad':"傷心",
   'surprise':"吃驚",
   'neutral':"中性"}
```

```python
r_d ={'asian':"亞洲人",
   'indian':"印度人",
   'black':"黑人",
   'white':"白人",
   'middle eastern':"中東人",
   'latino hispanic':"拉丁美洲人"}
```

```python
g_d = {'Woman':"女生", 'Man':"男生"}
```

```python
text = f"年紀: {age}歲 \n" + f"性別: {g_d[gender]} \n" + f"表情: {e_d[emotion]} \n" + f"人種: {r_d[race]}" 
print(text)
```
